```python
import pandas as pd
import glob
import os

# 1. 파일 패턴 설정
file_pattern = 'P*_ForceDiscrimination.csv'
file_list = glob.glob(file_pattern)

found_files = []
missing_files = []

# 결과 분석 시작
if not file_list:
    print("❌ [오류] 해당 패턴('P*_ForceDiscrimination.csv')과 일치하는 파일이 폴더에 없습니다.")
    print(f"현재 작업 경로: {os.getcwd()}")
else:
    for file_path in file_list:
        try:
            df = pd.read_csv(file_path)
            if 'Comparison' in df.columns:
                # 숫자 2.0 또는 문자열 '2.0' 개수 확인
                count = ((df['Comparison'] == 2.0) | (df['Comparison'] == '2.0')).sum()
                if count > 0:
                    found_files.append((file_path, count))
                else:
                    missing_files.append(file_path)
            else:
                missing_files.append(file_path)
        except Exception as e:
            print(f"파일 읽기 실패 ({file_path}): {e}")

    # 2. 최종 요약 보고 (이 부분을 확인하세요!)
    print("\n" + "="*60)
    print(f"📊 파일 분석 요약 보고서")
    print("="*60)
    print(f"1. 전체 분석 대상 파일 수: {len(file_list):>3}개")
    print(f"2. '2.0' 데이터 포함 파일 수: {len(found_files):>3}개")
    print(f"3. '2.0' 데이터가 없는 파일 수: {len(missing_files):>3}개")
    print("-" * 60)

    if found_files:
        print(f"✅ [상세 목록: '2.0'이 포함된 파일]")
        total_data_points = 0
        for f, c in found_files:
            print(f" - {f:<30} | {c:>3}개 발견")
            total_data_points += c
        print("-" * 60)
        print(f"📢 모든 파일에서 발견된 '2.0' 데이터 총합: {total_data_points}개")
    else:
        print("🔎 '2.0' 데이터가 포함된 파일이 하나도 없습니다.")
    print("="*60)
```
